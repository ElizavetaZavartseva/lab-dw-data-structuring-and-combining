## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_company_1 = pd.read_csv(url_1)

url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_company_2 = pd.read_csv(url_2)

url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_company_3 = pd.read_csv(url_3)
```

```python
insurance_company_1.head(10)
```

```python
insurance_company_3.head(10)
```

```python
print(f"insurance_company_1: rows - {insurance_company_1.shape[0]}, columns - {insurance_company_1.shape[1]}")
print(f"insurance_company_2: rows - {insurance_company_2.shape[0]}, columns - {insurance_company_2.shape[1]}")
print(f"insurance_company_3: rows - {insurance_company_3.shape[0]}, columns - {insurance_company_3.shape[1]}")
```

```python
#Solition_1 - Combining Data

insurance_company = pd.concat([insurance_company_1, insurance_company_2, insurance_company_3], axis=0)

print(f"insurance_company: rows - {insurance_company.shape[0]}, columns - {insurance_company.shape[1]}")

insurance_company

#Used method concat to combine 3 DataFrmes vertically.
```

```python
#Solution_2 - Cleaning Data

insurance_company.nunique()
```

```python
#Solution_2 - Cleaning Data: Cleaning Column Names

insurance_company.rename(columns= {'ST': 'state'}, inplace=True)
#Changed the name of colimn "state".

insurance_company.columns = [i.lower().replace(' ','_').strip() for i in insurance_company.columns]
#Change column names (repleced ' ' to '_' and applied lower case).

insurance_company
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

print(insurance_company.gender.value_counts())

insurance_company.gender = insurance_company['gender'].map({'Male': 'M', 'female': 'F', 'Femal':'F', 'M': 'M', 'F': 'F'})
print(insurance_company.gender.value_counts())
#Standardized column "gender"
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

print(insurance_company.state.value_counts())

insurance_company.state = insurance_company['state'].map({'Oregon': 'Oregon', 'California': 'California', 'Arizona':'Arizona', 'Cali': 'California', 'Nevada': 'Nevada', 'Washington': 'Washington', 'WA': 'Washington', 'AZ': 'Arizona'})
print(insurance_company.state.value_counts())
#Standardized column "state"
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

print(insurance_company.education.value_counts())

insurance_company.education = insurance_company.education.str.replace("Bachelors", "Bachelor")
print(insurance_company.education.value_counts())
#Standardized column "education"
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

print(insurance_company.customer_lifetime_value.value_counts())

insurance_company.customer_lifetime_value = insurance_company.customer_lifetime_value.str.replace("%", "")
print(insurance_company.customer_lifetime_value.value_counts())
#Standardized column "customer_lifetime_value"
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

print(insurance_company.vehicle_class.value_counts())

insurance_company.vehicle_class = insurance_company.vehicle_class.str.replace("Sports Car", "Luxury").str.replace("Luxury SUV", "Luxury").str.replace("Luxury Car", "Luxury")
print(insurance_company.vehicle_class.value_counts())
#Standardized column "vehicle_class"
```

```python
#Solution_2 - Cleaning Data: Cleaning Invalid Names

insurance_company.nunique()
```

```python
#Solution_2 - Formatting Data: 

print(insurance_company.dtypes)

insurance_company['customer_lifetime_value'] = insurance_company['customer_lifetime_value'].str.rstrip('%').astype('float') / 100
print(insurance_company.dtypes)
#customer_lifetime_value value should be numeric
```

```python
#Solution_2 - Formatting Data: 

print(insurance_company.number_of_open_complaints.unique())

new_number_of_open_complaints = insurance_company.number_of_open_complaints.apply(lambda x: x.split('/')[1] if isinstance(x, str) and x != "NaN" else x).astype(float)

insurance_company.number_of_open_complaints = new_number_of_open_complaints
print(insurance_company.number_of_open_complaints.unique())
#Changed the format of Number of open complaints. 
```

```python
#Solution_2 - Dealing with Null values

print(insurance_company.isna().sum())

insurance_company = insurance_company.dropna(how = 'all')
#First of all, I decided to delete rows that have all columns with null values.

insurance_company['customer_lifetime_value'].fillna(insurance_company['customer_lifetime_value'].mean(), inplace=True)
#Since this column has float64 datatype I made the decision to fill the null values with the column mean value.

insurance_company.gender.fillna("No gender", inplace = True)
#For column "gender" I went with filling null values as "No gender".

print(insurance_company.isna().sum())
#As evident now, in this dataset we don't have any null values.
```

```python
#Solution_2 - Dealing with duplicates

print(f"insurance_company: rows - {insurance_company.shape[0]}, columns - {insurance_company.shape[1]}")

print(f"Number of duplicates: {insurance_company.duplicated().sum()}")
#Here, we can see number of duplicates
```

```python
#Solution_2 - Dealing with duplicates

insurance_company.drop_duplicates(inplace=True)
print(f"Number of duplicates: {insurance_company.duplicated().sum()}")
print(f"insurance_company: rows - {insurance_company.shape[0]}, columns - {insurance_company.shape[1]}")
#Deleted all duplicates
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_sals_channel = insurance_company_new.pivot_table(index='sales_channel', values=['customer_lifetime_value'], aggfunc='sum')
pivot_sals_channel = pivot_sals_channel.round(2)
```

```python
pivot_sals_channel

#The Agent sales channel has the highest total customer lifetime value. 
#This suggests that customers acquired or managed through agents tend to generate more lifetime value, 
#indicating possibly higher customer satisfaction, loyalty, or more lucrative customer segments handled by agents.
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_avg_customer_lifetime_value = insurance_company_new.pivot_table(index='education', columns = ['gender'], values=['customer_lifetime_value'], aggfunc='mean')
pivot_avg_customer_lifetime_value = pivot_avg_customer_lifetime_value.round(2)
```

```python
pivot_avg_customer_lifetime_value

#Females in group "High School or Below" have the highest customer lifetime value overall ($8,675.22). This suggests that females with this level of education may have higher purchasing power or loyalty.
#Males with college education have a notably higher customer lifetime value ($8,052.46) compared to their female counterparts. This indicates that males in this category might engage more with products or services offered, leading to a higher lifetime value.
```
